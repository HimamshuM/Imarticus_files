#### Que 1 : Count all letters, digits, and special symbols from a given string
str1 = "I@#ma26rt^&ic5us"

```python
str1 = "I@#ma26rt^&ic5us"
lst = list(str1)
count_digit=0
count_letters = 0 
count_syb = 0 
for i in lst:
    if i.isnumeric()==True:
        count_digit+=1
    elif i.isalpha() ==True:
        count_letters+=1
    else: 
        count_syb+=1     
print("Number of digits in string: ",count_digit)
print("Number of letters in string: ",count_letters)
print("Number of special symbols in string: ",count_syb)
```

#### Que2 : Write a Python program to find numbers divisible by nineteen or thirteen from a list of numbers using Lambda.
a=[19, 65, 57, 39, 152, 639, 121, 44, 90, 190]

```python
# List of numbers
a = [19, 65, 57, 39, 152, 639, 121, 44, 90, 190]

# Filter numbers divisible by 19 or 13
result = list(filter(lambda x: x % 19 == 0 or x % 13 == 0, a))

# Print the result
print("Numbers divisible by 19 or 13:", result)
```

### Que 3: Find all of the words in a string that are less than 5 letters

### Que 4: Calculate income tax for the given income by adhering to the below rules
Expected Output:
For example, suppose the taxable income is 45000 the income tax payable is
10000 0% + 10000 10% + 25000 20% = $6000.

```python
def calculate_tax(income):
    if income <=10000:
        print("Not in tax slab")
    elif income>10000 and income<=20000: 
        tax = 10000*(10/100)
    else: 
        tax = 10000*(10/100) + (income - 10000-10000) * (20/100)
    return tax
    
```

```python
income = int(input("Enter your income here:"))
tax = calculate_tax(income)
print("yout payable tax is: ",tax)
```

### Que 5: Given below are the list of positive and negative words. Also a list of tweets is provided. Separate out the positive and negative comments¶
positive = ['good','awesome', 'best', 'nice']
negative = ['worst','awful', 'bad']
tweets = ['This government policies are good', 'bad implementation', 'The way he played showed that he is one of the best players in the world', 'Her acting in the play was awesome', 'The wine tastes awful', 'It's nice to hear this little kid's laugh']

```python
positive = ['good','awesome', 'best', 'nice'] 
negative = ['worst','awful', 'bad']
tweets = ['This government policies are good', 'bad implementation', 'The way he played showed that he is one of the best players in the world', 
          'Her acting in the play was awesome', 
          'The wine tastes awful',"It's nice to hear this little kid's laugh"]
possitive_comment = []
negative_comment = []

for tweet in tweets : 
    tweet = tweet.lower()
    if any(word in tweet for word in positive):
        possitive_comment.append(tweet)
    elif any(word in tweet for word in negative):
        negative_comment.append(tweet)

print("Possitive comment: ")
print("\n".join(possitive_comment))
print("\n Negative comment: ")
print("\n".join(negative_comment))        
```

### Que : 7 Load the iris dataset. 
Find only those records whose sepal_length = 5.1 and sepal_width = 3.5

```python
import pandas as pd
df = pd.read_csv(r"D:\[ Learning ] _ DataScience\Imarticus\iris.data.csv",header=None)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df
```
