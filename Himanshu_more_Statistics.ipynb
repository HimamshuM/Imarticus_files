Que 1:  From the pack of 52 cards, three cards are drawn randomly without replacement then what is the probability that one card is a diamond, one card is a heart and one is spade? 

	Choosing 3 cards from pack of card =   52C3 =  52*51*50 /  3*2*1  =22,100
1 card from the 13 diamonds = 13C1
1 card from the 13 hearts = 13C1 
1 card from the 13 spades = 13C1 

Therefore = 13 * 13 * 13 = 2197

	So probability = 2197 / 22,100

Probability = 0.0994

probability is approximately 0.0994, or about 9.94%.



Que 2. The information was gathered by a survey done by a production company: 42 % of the respondents said that they like action movies, 54 % like comedy movies, 36 % like drama movies, and 12 % like horror movies. If a person is selected at random, find the probability that his or her favourite movie type is either action or drama. solve this question

We are given the following probabilities:
•	P(Action)  =  42%=0.42
•	 P(Drama)  = 36%=0.36

P(Action or Drama)   =   P(Action) + P(Drama)
P(Action or Drama)	=  0.42  +  0.36  =  0.78

Que 3.  An MNC company receives 450 applications from applicants in one hour. Find the probability of 
a. Receiving 10 applications in 1 minute. 
b. Receiving at least 17 applications in 2 minutes.


	1. Calculate the rate of applications per minute:

there are 60 minutes in 1 hour, the average number of applications received per minute is:  
450 / 60 = 7.5 applications per minute
	P(X=10)  =  (7.5 ^  10 * e−7.5 ) / 10!   = 0.0858
	
The probability of receiving 10 applications in 1 minute is approximately 0.0858, or 8.58%.

	2. Receiving at least 17 applications in 2 minutes. 

the average number of applications per minute is 7.5   for 2 minutes :  7.5 * 2 = 15

Using the complement rule:  P(X >=  17 ) = 1 – P(X <= 16)
The probability of receiving at least 17 applications in 2 minutes is approximately 0.3359

a) The probability of receiving 10 applications in 1 minute is 8.58%.
b) The probability of receiving at least 17 applications in 2 minutes is 33.59%

Que 8. In the population, the mean weight is 82. A team of dietitians wants to test a new protein supplier to see if it has either a positive or negative effect on diet, or no effect at all. A sample of 15 participants who have taken the protein supplier. Did the protein supplier affect diet? Where alpha=0.05. 



Null Hypothesis H0: The protein supplier has no effect on diet (mean difference = 0).
Alternative Hypothesis H1: The protein supplier has an effect on diet (mean difference ≠ 0).
Set alpha level α=0.05
Test statistic: Use the paired t-test to calculate the t-statistic.
The P -value is : 0.0490 
	As P-Value is lower than 0.05 we will reject the null Hypothesis. 
	The protein supplier has an effect on diet.
