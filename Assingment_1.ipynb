```python
# importing libraries

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mob = pd.read_csv('trainRF.csv')
mob.head()
```

### Analysis - Data Cleaning

```python
mob.isnull().sum()
```

- There is no null present in Dataset, so we are good to analyse the data

```python
mob['battery_power'].corr(mob['price_range'])
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(mob.corr().round(2),annot=True)
```

- From above heatmap we can see battery power, px_hight,px_width are having good coorelation with price range but,
ram is higly correlated with price range. 

### Price Range Analysis with ram

```python
# plt.figure(figsize=(10,5))
# plt.bar(mob['price_range'],mob['ram'], color = 'purple')
sns.barplot(x=mob['price_range'],y=mob['ram'],data=mob,color = 'purple')
mob['price_range'] = mob['price_range'].replace({0: 'Low', 1:'Medium', 2:'High', 3: 'Expencive'})
plt.xlabel('Price Range')
plt.ylabel("Ram size in Mb")
plt.title("Ram and Price Comparision")
plt.show()
```

- From Above graph we can conclude that by incresing Ram size, mobile price going to be increase.
